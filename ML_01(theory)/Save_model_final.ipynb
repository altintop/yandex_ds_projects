```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# подключить библиотеку
import joblib
```

```python
# обучить модель (как раньше)
df = pd.read_csv('train_data_final.csv')

df_train, df_valid = train_test_split(df, test_size=0.25, random_state=12345)

features_train = df_train.drop(['last_price'], axis=1)
target_train = df_train['last_price'] / 1000000
features_valid = df_valid.drop(['last_price'], axis=1)
target_valid = df_valid['last_price'] / 1000000

# Модель решающего дерева
all_rmse = []
for i in range(1,21):
    model = DecisionTreeRegressor(random_state=12345, max_depth=i)
    model.fit(features_train, target_train)
    predicted_valid = model.predict(features_valid)
    rmse = mean_squared_error(target_valid, predicted_valid)**0.5
    all_rmse.append(rmse)
    print('DecisionTreeRegressor, max_depth=', i, ', rmse = ', rmse)
print(min(all_rmse), '\n')
# Вывод: Самый лучший DecisionTreeRegressor(random_state=12345, max_depth=10), rmse = 6382083.993512222
print('----------------------------------------------------------------')

# Модель случайного леса
all_rmse = []
for d in range(1, 11):
    for estim in range(10, 51, 10):
        model = RandomForestRegressor(n_estimators=estim, max_depth=d, random_state=12345)
        model.fit(features_train, target_train)
        predicted_valid = model.predict(features_valid)
        rmse = mean_squared_error(target_valid, predicted_valid)**0.5
        all_rmse.append(rmse)
        print('RandomForestRegressor, n_estimators=', estim, ', max_depth =', d, ', rmse = ', rmse)
print(min(all_rmse), '\n')
# Вывод: Самый лучший RandomForestRegressor, n_estimators= 40 , rmse =  6916870.747848356
print('----------------------------------------------------------------')

# Модель линейной регрессии
model = LinearRegression()
model.fit(features_train, target_train)
predicted_valid = model.predict(features_valid)
rmse = mean_squared_error(target_valid, predicted_valid)**0.5
print('LinearRegression, rmse = ', rmse)
```

```python
# В итоге самая лучшая модель с минимальным rmse - это DecisionTreeRegressor(random_state=12345, max_depth=10)
df = pd.read_csv('train_data_final.csv')
features = df.drop(['last_price'], axis=1)
target = df['last_price'] / 1000000

model = RandomForestRegressor(n_estimators=40, max_depth=10, random_state=12345)
model.fit(features, target)
# сохранить модель
# первый аргумент - модель
# второй аргумент - путь к файлу
joblib.dump(model, 'final_model.joblib')
```
